```python
import FinanceDataReader as fdr #pip install -U finance-datareader 로 설치하시면 됩니다.
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

start_date = '2017-01-01'
end_date = '2024-01-01'

kodex200 = fdr.DataReader('069500', start=start_date, end=end_date) #번호 검색해서 바꾸

etf_df = pd.DataFrame()

etf_df['KODEX_200'] = kodex200['Close']

plt.figure(figsize=(12, 6))
plt.plot(etf_df.index, etf_df['KODEX_200'], label='KODEX 200')
plt.xlabel('날짜')
plt.ylabel('종가')
plt.title('7년간 주식그래프') #변경가능합니다
plt.legend()
plt.grid(True)
plt.show()
```
