```python
import FinanceDataReader as fdr
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
import matplotlib as mpl
import yfinance as yf
warnings.filterwarnings('ignore')
mpl.rc('font', family='Malgun Gothic')  
mpl.rcParams['axes.unicode_minus'] = False

start_date = '2005-01-01' #경선확정일 - 10
end_date = '2025-12-31' #경선확정일 + 10



thema1 = yf.download('013000.KS', start=start_date, end=end_date)  # KODEX 200


thema= pd.DataFrame()

thema['thema1'] = thema1['Close']


plt.figure(figsize=(12, 6))
plt.plot(thema.index, thema['thema1'], label='바른손')


#경선 일에 빨간 점선 줄
plt.axvline(x=pd.to_datetime('2012-06-01'), color='red', linestyle='--', label='경선일') #경선일을 수정하려면 datatime변수 수정해주시면 됩니다.

plt.xlabel('날짜')
plt.ylabel('종가')
plt.title('19대_문재인_테마주')
plt.legend()
plt.grid(True)
plt.show()
```
