```python
import FinanceDataReader as fdr
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
import matplotlib as mpl
import matplotlib.dates as mdates
warnings.filterwarnings('ignore')
mpl.rc('font', family='Malgun Gothic')  
mpl.rcParams['axes.unicode_minus'] = False

start_date = '2025-04-23' # - 10
end_date = '2025-05-13' # + 10

#2      .# thema2 = fdr.DataReader('015260', start=start_date, end=end_date)
#(1)      plt      
thema1 = fdr.DataReader('090080', start=start_date, end=end_date) #1
thema2 = fdr.DataReader('010770', start=start_date, end=end_date)
thema3 = fdr.DataReader('014160', start=start_date, end=end_date)
thema4 = fdr.DataReader('025750', start=start_date, end=end_date)


thema= pd.DataFrame()
#(1)     
thema['thema1'] = thema1['Close']
thema['thema2'] = thema2['Close']
thema['thema3'] = thema3['Close']
thema['thema4'] = thema4['Close']



#(1)     
plt.figure(figsize=(12, 6))
plt.plot(thema.index, thema['thema1'], label='')
plt.plot(thema.index, thema['thema2'], label='')
plt.plot(thema.index, thema['thema3'], label='')
plt.plot(thema.index, thema['thema4'], label='')
plt


#    
plt.axvline(x=pd.to_datetime('2025-05-03'), color='red', linestyle='--', label='') #  datatime  .

plt.xlabel('')
plt.ylabel('')
plt.title('21__')
plt.legend()
plt.grid(True)
plt.show()


label_dict = {
    'thema1': '',
    'thema2': '',
    'thema3': '',
    'thema4': '',
}

#    
thema_renamed = thema.rename(columns=label_dict)

# CSV  (index  )
thema_renamed.to_csv('21____.csv', encoding='utf-8-sig')

```

```python
#  
#    
daily = thema.resample('d').last()
d_df = daily.pct_change()
daily_log_returns = d_df.apply(lambda x: np.log(1 + x)).cumsum() #  

#  
plt.figure(figsize=(12, 6))
plt.axhline(y=0.0, color='red', linestyle='--')  #  0 

#   
columns = ['thema1', 'thema2', 'thema3', 'thema4']
labels = ['','','','']  #   .
colors = plt.cm.tab10.colors

for i, (col, label) in enumerate(zip(columns, labels)):
    plt.plot(daily_log_returns.index, daily_log_returns[col], label=label, color=colors[i % 10])

    last_date = daily_log_returns.index[-1]
    log_return = daily_log_returns[col].iloc[-1]
    normal_return = np.exp(log_return) - 1  #       

    plt.text(last_date + pd.Timedelta(days=1),  
             log_return,
             f'{normal_return * 100:.1f}%',   
             fontsize=11,
             va='center',
             ha='left',
             color=colors[i % 10])
    
plt.xlim(daily_log_returns.index[0], daily_log_returns.index[-1]) 

plt.xticks(fontsize=9, rotation=55)

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=2)) #   y    ,   8    .
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.yticks(fontsize=7)

plt.xlabel('', fontsize=13)
plt.ylabel('', fontsize=13)
plt.title('21___')
plt.legend(fontsize=10)
plt.grid(True)

plt.tight_layout()
plt.show()

label_dict = {
    'thema1': '',
    'thema2': '',
    'thema3': '',
    'thema4': '',
}



#     
daily_log_returns_named = daily_log_returns.rename(columns=label_dict)

#       
daily_normal_returns = np.exp(daily_log_returns_named) - 1

#   
daily_normal_returns_percent = daily_normal_returns * 100

# 
daily_normal_returns_percent.to_csv('21___.csv', encoding='utf-8-sig')
```
